**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
movie_data['profit'] = movie_data['revenue'] - movie_data['budget']
movie_profit = pd.DataFrame(movie_data.groupby('release_year').revenue.agg('mean')).reset_index()
```

```python
fig2 = plt.figure(figsize = (20, 5))
plt.errorbar(data = movie_profit, x = 'release_year', y = 'revenue')

plt.xticks(rotation = 45)

plt.xlabel('Year', fontsize = 20)
plt.ylabel('Revenue', fontsize = 20)

plt.title('Revenue by Year', fontsize = 20);
```

由上图可知，随着年份的变化，电影净利润发生了较大的起伏变化：<br>
1、70年代中期，电影净利润发生了剧烈波动，即出现了历史最高峰，也出现了低谷；<br>
2、除70年代中期，60年代初期、70年代中期、80年代末期，电影净利润均处于低谷；<br>
3、除70年代中期，80年代初中期、90年代末期，电影净利润较高；<br>
4、进入20世纪，电影净利润有所下降，但保持平稳。

```python
index_list = list(top_directors_df1['level_1'])
```

```python
top_directors_df = top_directors_df.loc[index_list]
top_directors_df = top_directors_df.groupby(['director']).head(3)
```

```python
rank_list = ['Rank1', 'Rank2', 'Rank3'] * 10
top_directors_df['rank'] = rank_list
top_directors_df = top_directors_df[['director', 'revenue', 'rank']]
```

```python
top_directors_df.plot(kind = 'bar', x = 'director', figsize = (20,5),  fontsize = 10, rot = 45);
```

**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
movie_data['release_date1'] = pd.to_datetime(movie_data['release_date'])
```

```python
movie_data['release_month'] = [i.month for i in movie_data['release_date1']]
```

```python
year_list = list(range(1968,2006))
```

```python
jun_month = movie_data[(movie_data.release_month == 6) & (movie_data.release_year.isin(year_list))]
```

```python
jun_month1 = jun_month[['original_title', 'release_year']].groupby('release_year').count()
```

```python
sb.pointplot(data = jun_month1, x = jun_month1.index, y = 'original_title' )
plt.xticks(rotation = 45);
plt.title('Jun Movie Amount');
```

由上图可知，1968年-2015年六月电影的数量经历了两次明显变化：<br>
1、80年代初期，电影数量经历了一次提升，到90年代初期回落；<br>
2、进入20世纪，电影数量经历了快速提升。

**[选做]任务3.5：**分析1968年~2015年六月电影 `Comedy` 和 `Drama` 两类电影的数量的变化。

```python
genres_list = list(jun_month['genres'])
drama_list = []
comedy_list = []
for i in genres_list:
    if 'Drama' in i and 'Comedy' in i:
        drama_list.append(1)
        comedy_list.append(1)
    elif 'Drama' in i:
        drama_list.append(1)
        comedy_list.append(0)
    elif 'Comedy' in i:
        comedy_list.append(1)
        drama_list.append(0)
    else:
        comedy_list.append(0)
        drama_list.append(0)   
```

```python
genres_df = pd.DataFrame({'drama': drama_list, 'comedy': comedy_list} )
```

```python
jun_month = jun_month[['release_year']].reset_index()
jun_month = jun_month['release_year']
```

```python
genres_df = pd.concat([jun_month, genres_df],axis = 1)
```

```python
genres_df = genres_df.groupby('release_year').count()
```

```python
fig3 = plt.figure(figsize = (20,10))
width = 0.3
plt.bar(x = genres_df.index - width/2, height = genres_df.drama, width = width, color = 'blue', label = 'drama')
plt.bar(x = genres_df.index + width/2, height = genres_df.comedy, width = width, color = 'yellow', label = 'comedy')

plt.xticks(range(1968,2015), rotation = 45, fontsize = 10)
plt.yticks(fontsize = 15)

plt.xlabel('Release Year', fontsize = 20)
plt.ylabel('Amount', fontsize = 20)

plt.title('Drama and Comedy every Jun.', fontsize = 20);
plt.legend()
```

由上图可知，1968年-2015年六月电影 Comedy 和 Drama 两类电影的数量有以下特点：<br>
1、两类电影数量完全一致；<br>
2、进入20世纪，两类电影数量都有了快速发展,但随后又回落；<br>
3、1968年-2015年期间，大多数年份的六月，都没有两种电影类型。
