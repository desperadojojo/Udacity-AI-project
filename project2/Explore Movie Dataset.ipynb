**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
movie_data['profit'] = movie_data['revenue'] - movie_data['budget']
movie_profit = pd.DataFrame(movie_data.groupby('release_year').revenue.agg('mean')).reset_index()
```

```python
fig2 = plt.figure(figsize = (20, 5))
plt.errorbar(data = movie_profit, x = 'release_year', y = 'revenue')

plt.xticks(rotation = 45)

plt.xlabel('Year', fontsize = 20)
plt.ylabel('Revenue', fontsize = 20)

plt.title('Revenue by Year', fontsize = 20);
```

由上图可知，随着年份的变化：<br>
1、电影净利润整体呈现上升趋势；<br>
2、进入20世纪，电影净利润趋于平稳，并且略有下降。

**[选做]任务3.3：**选择最多产的10位导演（电影数量最多的），绘制他们排行前3的三部电影的票房情况，并简要进行分析。

进行代码优化，用堆积图更好地展示数据

```python
director_data = movie_data[['director', 'original_title', 'revenue']]
```

```python
top10_directors = director_data.groupby('director').count()['original_title'].nlargest(10).index
#nlargest().index   numpy找出n个最大的元素的index
```

```python
top10_table = director_data[director_data['director'].isin(top10_directors)]
```

```python
top10_table = top10_table.sort_values(['director','revenue'], ascending=[False,False]).set_index(['director','original_title']) 
```

```python
top10_table.groupby(level=0).head(3)
```

```python
top10_table.groupby(level=0).head(3).unstack().plot(kind='bar',stacked=True, legend=False, rot=30, figsize=(10,5)) #stacked 堆积
plt.ylabel('revenue')
plt.title("Top 10 prolific directors with 3 movies of highest revenues");
```

由上图可知，最多产的10位导演排行前3的三部电影的票房有较大差异：<br>
1、票房成绩最高的导演Spielberg的票房大约是最低的Allen的约10倍；<br>
2、Spielberg也是单部成绩最高的导演；<br>
3、除最高和最低成绩两位导演，剩下8位导演票房成绩分布在0.5-1.5区间内。

**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
movie_data['release_date1'] = pd.to_datetime(movie_data['release_date'])
```

```python
movie_data['release_month'] = [i.month for i in movie_data['release_date1']]
```

```python
year_list = list(range(1968,2006))
```

```python
jun_month = movie_data[(movie_data.release_month == 6) & (movie_data.release_year.isin(year_list))]
```

```python
jun_month1 = jun_month[['original_title', 'release_year']].groupby('release_year').count()
```

```python
plt.figure(figsize = (20,10))
sb.pointplot(data = jun_month1, x = jun_month1.index, y = 'original_title')

plt.xticks(rotation = 45, fontsize = 15);
plt.yticks(fontsize = 15)

plt.xlabel('Release Year', fontsize = 20)
plt.ylabel('Amount', fontsize = 20)

plt.title('Jun Movie Amount', fontsize = 20);
```

由上图可知，1968年-2015年六月电影的数量整体处于递增趋势：<br>
1、80年代初期，电影数量经历了一次快速提升，到80年代末期回落；<br>
2、进入20世纪，电影数量经历了再一次快速提升。

**[选做]任务3.5：**分析1968年~2015年六月电影 `Comedy` 和 `Drama` 两类电影的数量的变化。

```python
genres_series = pd.Series(jun_month['genres'])
```

```python
jun_month['drama'] = genres_series.str.contains('Drama').astype('int')
jun_month['comedy'] = genres_series.str.contains('Comedy').astype('int')
```

```python
genres_df = jun_month[['release_year', 'drama', 'comedy']]
genres_df = genres_df.groupby('release_year').count()
```

```python
genres_df.plot(kind='bar',legend=True, rot=30, figsize=(20,5))
plt.title('Drama and Comedy amount every Jun');
```

由上图可知，1968年-2015年六月电影 Comedy 和 Drama 两类电影的数量有以下特点：<br>
1、两类电影数量完全一致；<br>
2、整体呈现上升趋势；<br>
3、80年代初期上升至中期达到峰值，80年底末期回落；<br>
4、90年代初期开始持续上升。
